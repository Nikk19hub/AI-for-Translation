```python
from transformers import MarianMTModel, MarianTokenizer
import re

def clean_text(text):
    """ Basic text cleaning function """
    text = text.strip()
    text = re.sub(r"\s+", " ", text)  # Remove extra whitespaces
    return text

def translate_text(text, model_name='Helsinki-NLP/opus-mt-en-de', num_beams=5):
    text = clean_text(text)
    if not text:
        return "Input text is empty. Please enter some text to translate."

    try:
        # Load the tokenizer and model
        tokenizer = MarianTokenizer.from_pretrained(model_name)
        model = MarianMTModel.from_pretrained(model_name)

        # Tokenize the input text
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

        # Generate translation using beam search
        translated_tokens = model.generate(**inputs, num_beams=num_beams, early_stopping=True)

        # Decode the translated tokens
        translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

        return translated_text
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
input_text = input("Enter text in English: ")
translated_text = translate_text(input_text)
print(f"Translated text: {translated_text}")
```

```python
from PyPDF2 import PdfReader
from googletrans import Translator
from fpdf import FPDF

def read_pdf(file_path):
    pdf = PdfReader(file_path)
    text = ''
    for page_num in range(len(pdf.pages)):
        page = pdf.pages[page_num]
        text += page.extract_text()
    return text

def translate_text(text, src_lang='en', dest_lang='de'):
    translator = Translator()
    translated = translator.translate(text, src=src_lang, dest=dest_lang)

    if translated is None:
        print("Translation failed. Check network connection and API limits.")
        return ""  # Return an empty string to avoid further errors

    return translated.text

def create_pdf(text, output_path):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, text)
    pdf.output(output_path)

def translate_pdf(input_path, output_path):
    text = read_pdf(input_path)
    translated_text = translate_text(text)
    create_pdf(translated_text, output_path)

# Example usage (unchanged)
input_pdf_path = '/content/RANGE - ZT 15-22VSD (2023) - Antwerp (APF) - 2987 2030 00=1=Model Book=EN (1) (1).pdf'
output_pdf_path = 'translated_output.pdf'
translate_pdf(input_pdf_path, output_pdf_path)
```

```python
import pdfplumber
from transformers import MarianMTModel, MarianTokenizer
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter

# Function to translate text
def translate_text(text, model, tokenizer):
    translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True))
    return [tokenizer.decode(t, skip_special_tokens=True) for t in translated]

# Initialize the MarianMT model and tokenizer for English to German translation
model_name = 'Helsinki-NLP/opus-mt-en-de'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Open the PDF file
input_pdf_path = '/content/RANGE - ZT 15-22VSD (2023) - Antwerp (APF) - 2987 2030 00=1=Model Book=EN (1) (1)-1-5.pdf'
output_pdf_path = '/mnt/data/Translated_Document.pdf'

with pdfplumber.open(input_pdf_path) as pdf:
    c = canvas.Canvas(output_pdf_path, pagesize=letter)
    width, height = letter

    for page_num, page in enumerate(pdf.pages):
        # Extract text and layout information
        text = page.extract_text(x_tolerance=2, y_tolerance=2)
        words = page.extract_words()

        # Translate each word keeping its position
        translated_words = []
        for word in words:
            translated_text = translate_text(word['text'], model, tokenizer)
            translated_words.append((translated_text[0], word['x0'], word['top']))

        # Draw the translated text at the same positions in the new PDF
        for text, x, y in translated_words:
            c.drawString(x, height - y, text)

        c.showPage()

    c.save()

print("Translation completed and saved to", output_pdf_path)
```

# **Updated Marain NMT (1)**

```python
from transformers import MarianMTModel, MarianTokenizer
import re
import torch

def clean_text(text):
    """ Advanced text cleaning function """
    text = text.strip()
    text = re.sub(r"\s+", " ", text)  # Remove extra whitespaces
    text = re.sub(r"[^\w\s,.!?]", "", text)  # Remove special characters except punctuation
    return text

def translate_text(text, model_name='Helsinki-NLP/opus-mt-en-de', num_beams=5, length_penalty=1.0, max_length=512):
    text = clean_text(text)
    if not text:
        return "Input text is empty. Please enter some text to translate."

    try:
        # Load the tokenizer and model
        tokenizer = MarianTokenizer.from_pretrained(model_name)
        model = MarianMTModel.from_pretrained(model_name)

        # Tokenize the input text
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=max_length)

        # Generate translation using beam search
        translated_tokens = model.generate(
            **inputs,
            num_beams=num_beams,
            length_penalty=length_penalty,
            early_stopping=True,
            max_length=max_length
        )

        # Decode the translated tokens
        translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

        # Post-process the translated text
        translated_text = re.sub(r"\s([?.!,'](?:\s|$))", r"\1", translated_text)  # Fix spacing before punctuation

        return translated_text
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
input_text = input("Enter text in English: ")
translated_text = translate_text(input_text)
print(f"Translated text: {translated_text}")
```
